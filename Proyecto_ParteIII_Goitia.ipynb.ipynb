### **Caracterisiticas Clave**

**movie_id**: es el identificador único para cada película cuya estructura seria la siguiente tomando un ejemplo: tt0099088 es el indetificador de la pelicula "Back to the Future Part III", donde tt significa que es una entidad de título y 0099088 indica de forma única a "Back to the Future Part III".

**title**: titulo de la pelicula.

**year**:  año en que fue estrenada la pelicula.

**trivia**: sinopsis de la pelicula.

**mpaa**: clasificacion de la pelicula basado en los codigos de la Motion Picture Association of America.

**release_date**: fechan de lanzamiento.

**run_time**: tiempo de duracion de la pelicula.

**distributor**: estudio o casa productora encargada de la distribucion de la pelicula.

**director**: el encargado de dirigir la pelicula en su realizacion.

**writer**: escritor o guionista.

**producer**: productor.

**composer**: compositor.

**cinematographer**: cinematografo o el director de fotografia.

**main_actor_1**: actor principal.

**main_actor_2**: actor principal 2.

**main_actor31**: actor principal 3.

**main_actor_4**: actor principal 4.

**budget**: presupuesto que tuvo la produccion.

**domestic**: taquilla o recaudacion en el mercado domestico.

**international**: taquilla o recaudacion en el mercado internacional.

**worldwide**: taquilla o recaudacion en el mercado mundial.

**genre_1**: genero de la pelicula.

**genre_2**: genero de la pelicula 2.

**genre_3**: genero de la pelicula 3.

**genre_4**: genero de la pelicula 4.

**html**: enlace donde esta ubicada la pelicula dentro de la pagina https://www.boxofficemojo.com/.

```python
# tenemos en la columna "writer" un total de 9 datos nulos
nulos_writer = df_boxoffmoj[df_boxoffmoj['writer'].isnull()]

# seleccionamos las columnas específicas que nos interesan para visualizar
resultados = nulos_writer[['movie_id', 'title','year', 'writer']]

print(resultados)
```

```python
# diccionario con los índices de las filas y sus respectivos escritores
escritores = {
    15: "Jenny Livingston",
    41: "Madonna",
    1385: "Brian Herzlinger",
    1532: "Robert Greenwald",
    1651: "Guido Verweyen",
    2580: "Jon M. Chu",
    2750: "Dan Cutforth",
    2769: "Richard Everts",
    2833: "Morgan Spurlock",
}

# actualizamos la columna 'writer' 
for indice, escritor in escritores.items():
    df_boxoffmoj.loc[indice, 'writer'] = escritor

# Vverificamos que los datos nulos han sido llenados
print(df_boxoffmoj.loc[df_boxoffmoj['writer'].isnull(), ['movie_id', 'title', 'writer']])
```

```python
# tenemos las columnas "domestic" con 19 datos nulos, "international" con 410 y "worldwide" con 7
# estas corresponden a datos numericos acerca de la recaudacion en taquilla
# vamos a rellenar estos datos aplicando la mediana  
df_boxoffmoj['domestic'] = df_boxoffmoj['domestic'].fillna(df_boxoffmoj['domestic'].median())
df_boxoffmoj['international'] = df_boxoffmoj['international'].fillna(df_boxoffmoj['international'].median())
df_boxoffmoj['worldwide'] = df_boxoffmoj['worldwide'].fillna(df_boxoffmoj['worldwide'].median())
```

```python
# y finalmente en nuestra revision de valores nulos con encontramos con las columnas 
# genre_1, genre_2, genre_3, genre_4 referentes al genero de cada pelicula.
# en este caso combinaremos las columnas en una sola y asi tendremos mas condensada parte de la informacion.           
df_boxoffmoj['genres'] = df_boxoffmoj[['genre_1', 'genre_2', 'genre_3', 'genre_4']].apply(lambda x: ', '.join(x.dropna()), axis=1)

# eliminamos las columnas antiguas de géneros
df_boxoffmoj = df_boxoffmoj.drop(columns=['genre_1', 'genre_2', 'genre_3', 'genre_4'])
```

```python
# volvemos a visualizar la sumatoria de valores nulos para verificar que todo esta correcto
df_boxoffmoj.isnull().sum()
```

```python
# ahora eliminamos la columna "trivia", ya que no es una columna que vayamos a utilizar
df_boxoffmoj = df_boxoffmoj.drop(columns=['trivia'])

```

```python
# Identificamos duplicados en el dataset
num_duplicados= df_boxoffmoj.duplicated().sum()
print("La cantidad de duplicados es:", num_duplicados)
```

A continuacion realizaremos un cambio dentre del tipo de datos en el caso de las columnas budget, domestic, international y worlwide que, tienen un tipo de datos float y lo vamos a cambiar a int64, esto con el fin de optimizar mejor nuestros datos al momento de realiza futuras operaciones.

```python
df_boxoffmoj['budget']=df_boxoffmoj['budget'].astype('int64')
df_boxoffmoj['domestic']=df_boxoffmoj['domestic'].astype('int64')
df_boxoffmoj['international']=df_boxoffmoj['international'].astype('int64')
df_boxoffmoj['worldwide']=df_boxoffmoj['worldwide'].astype('int64')
```

Para profundizar unn poco mas en el tema de presupuestos y generos realizaremos un pequeño analisis estadistico haciendo una prueba ANOVA:

```python
from scipy.stats import f_oneway

genre_groups = [df_boxoffmoj_genre[df_boxoffmoj_genre['genre_1'] == genre]['budget'] for genre in df_boxoffmoj_genre['genre_1'].unique()]
stat, p_val = f_oneway(*genre_groups)
print('p-valor:', p_val)
```

De acuerdo con el resultado obtenido, tenemos un p-valor de 1.78, lo cual nos indica que existe una brecha entre los generos en peliculas y sus presupuestos, esto nos hace rechazar la hipotesis nula que nos dice lo contrario y aceptamos la hipotesis alternativa.

Analisis de Ratio

Ahoras analizaremos el ratio utilizando las columnas budget y domestic para determinar cuanto representa la recaudacion con respecto al presupuesto de la pelicula.

```python
# calculamos una nueva columna 
df_boxoffmoj_genre['ratio_domestic_budget'] = (df_boxoffmoj_genre['domestic'] / df_boxoffmoj_genre['budget']) * 100

# calculamos el ratio promedio por género
average_ratio_by_genre = df_boxoffmoj_genre.groupby('genre_1')['ratio_domestic_budget'].mean().sort_values(ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
average_ratio_by_genre.plot(kind='bar', color="yellow")
plt.xlabel('Género')
plt.ylabel('Ratio Promedio (Domestic / Budget)')
plt.title('Ratio Promedio por género')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

El genero de horror se planta como el primero en este grafico dejando en claro una gran eficacia en cuanto a la cantidad de recaudacion por encima de su presupuesto, esto se podria explicar plantando que son peliculas muy populares y rentables con una tasa de inversion moderada o baja en comparacion con otros filmes; lo cual nos lleva a un poder de rentabilidad inmenso pero, tomemos en cuenta que este genero tiene una cantidad que no supera las 200 peliculas por lo tanto el resultado de este grafico no debe tomarse como un valor absoluto y representativo.
